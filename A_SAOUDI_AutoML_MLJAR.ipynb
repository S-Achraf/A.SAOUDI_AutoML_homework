```python
import pandas as pd

print("1. Loading dataset...")
df = pd.read_csv("/content/homework.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
target = "price"   # change if needed

X = df.drop(columns=[target])
y = df[target]
```

```python
from supervised.automl import AutoML

print("\n2. Running MLJAR AutoML...")

automl = AutoML(
    mode="Explain",        # best for homework
    eval_metric="rmse",
    total_time_limit=300,  # 5 minutes
    algorithms=["Linear", "Random Forest", "LightGBM", "Xgboost"]
)

automl.fit(X, y)

```

```python
print("\n3. Leaderboard:")
automl.get_leaderboard()
```

```python
print("\n4. Predictions sample:")
predictions = automl.predict(X.head(10))
predictions
```

```python
print("\n5. Model summary:")
automl.report()
```
