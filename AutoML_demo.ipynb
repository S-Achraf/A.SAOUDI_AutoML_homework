```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
```

```python
print("1. Loading dataset...")
df = pd.read_csv("/content/homework.csv")

print("Dataset shape:", df.shape)
df.head()
```

```python
target = "price"   # CHANGE if your target column is different
print("\n2. Initializing H2O...")
h2o.init()

```

```python
h2o_df = h2o.H2OFrame(df)
print("\n3. Setting up AutoML...")

x = h2o_df.columns
y = target
x.remove(y)

print("Number of features:", len(x))
print("Target column:", y)
```

```python
train, test = h2o_df.split_frame(ratios=[0.8], seed=42)

print("Training rows:", train.shape[0])
print("Testing rows:", test.shape[0])
```

```python
print("\n4. Running AutoML...")

aml = H2OAutoML(
    max_models=10,
    max_runtime_secs=300,
    seed=42,
    verbosity="info"
)

aml.train(x=x, y=y, training_frame=train)
```

```python
print("\n5. Leaderboard:")
aml.leaderboard.head()
```

```python
best_model = aml.leader
print("Best model ID:", best_model.model_id)
```

```python
print("\n6. Predictions sample:")
predictions = best_model.predict(test)
predictions.head()
```

```python
print("\n7. Model Performance:")
performance = best_model.model_performance(test)
performance
```

```python
print("\n8. Shutting down H2O...")
h2o.cluster().shutdown()
```
